```python
# libraries
import numpy as np # used for handling numbers
import pandas as pd # used for handling the dataset
```

```python
dataset = pd.read_csv('output_file.csv')
dataset.head()
```

```python
print("missing values")
print(dataset.isnull().sum())
print("")
```

```python
print(dataset['x'])
```

```python
dataset = dataset.rename(columns={'x' : 'x_min', 'y' : 'y_min'})
dataset.head(10)
```

```python
dataset['x_max'] = dataset['x_min'] + dataset['width']
dataset['y_max'] = dataset['y_min'] + dataset ['height']
dataset.head(10)
```

```python
bucket_name = 'leaf_data_final'
gcs_path = f'gs://{bucket_name}/'
dataset['gcs_path'] = gcs_path + dataset['file_name']
dataset['head'] = 'Training'
dataset.head(10)
```

```python
# X_MIN,Y_MIN,X_MAX,Y_MIN, X_MAX,Y_MAX,X_MIN,Y_MAX,
df2 = dataset[['head', 'gcs_path', 'Type', 'x_min', 'y_min', 'x_max', 'y_min','x_max','y_max','x_min', 'y_max']]
df2.head(10)
```

```python
df2.dtypes
```

```python
columns_to_normalize = ['x_min', 'x_max', 'y_min', 'y_max']

# Perform normalization on the selected columns
df2[columns_to_normalize] = (df2[columns_to_normalize] - df2[columns_to_normalize].min()) / (df2[columns_to_normalize].max() - df2[columns_to_normalize].min())

# Check the updated DataFrame
print(df2.head())
```

```python
print("missing values")
print(df2.isnull().sum())
print("")
```

```python
df2.to_csv('leaf_defect_final.csv', index=False, header=False)
```
